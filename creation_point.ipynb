```python
import pandas as pd
from geopy.geocoders import Nominatim
loc = Nominatim(user_agent="GetLoc")
import json
import random
from shapely.geometry import Point, Polygon
```

```python
def create_10000_points(lat,long,poly_New_Caledonia):
    liste_points = []
    while len(liste_points) < 10000:
        x = random.random()
        if x <= 0.4:
            longitude = long + random.uniform(-0.05,0.05)
            latitude = lat + random.uniform(-0.05,0.05)
        elif 0.4 < x <= 0.75:
            longitude = long + random.uniform(-0.10,0.10)
            latitude = lat + random.uniform(-0.10,0.10)
        elif 0.75 < x <= 1:
            longitude = long + random.uniform(-0.20,0.20)
            latitude = lat + random.uniform(-0.20,0.20)
        point = Point(longitude,latitude)
        for i in range(len(poly_New_Caledonia['coordinates'])):
            if point.within(Polygon(poly_New_Caledonia['coordinates'][i][0])):
                liste_points.append([latitude,longitude])
    return liste_points

def create_10000_points_other(lat,long,poly_New_Caledonia):
    liste_points = []
    while len(liste_points) < 10000:
        x = random.random()
        if x <= 0.4:
            longitude = long + random.uniform(-0.5,0.5)
            latitude = lat + random.uniform(-0.5,0.5)
        elif 0.4 < x <= 0.75:
            longitude = long + random.uniform(-0.10,0.10)
            latitude = lat + random.uniform(-0.10,0.10)
        elif 0.75 < x <= 1:
            longitude = long + random.uniform(-0.20,0.20)
            latitude = lat + random.uniform(-0.20,0.20)
        point = Point(longitude,latitude)
        for i in range(len(poly_New_Caledonia['coordinates'])):
            if point.within(Polygon(poly_New_Caledonia['coordinates'][i][0])):
                liste_points.append([latitude,longitude])
    return liste_points
```

```python
df_points = pd.read_csv(r"C:\Users\33638\Downloads\population-legale-de-la-nouvelle-caledonie.csv",sep=";")
```

```python
with open(r"C:\Users\33638\Desktop\Pacific Data Hub\New Calédonia\metropole-et-outre-mer.geojson") as handle:
    country_geo = json.loads(handle.read())
```

```python
poly_New_Caledonia = {'type': 'MultiPolygon',
 'coordinates': country_geo["coordinates"][40:83]}
```

```python
df = df_points[df_points.column_6==2019][["commune","column_2"]]
```

```python
df["column_2"] = df["column_2"] / df["column_2"].sum() * 0.9
```

```python
df["loc"] = None
for row in df.index:
    df["loc"][row] = [loc.geocode(f"{df.commune[row]},New calédonia")[1][0],loc.geocode(f"{df.commune[row]},New calédonia")[1][1]]
```

```python
df.reset_index(inplace=True,drop=True)
df.loc[33] = ["Other",0.100000,[-21.2, 165.5]]
```

```python
df["points"] = None

for row in df[:33].index:
    df.points[row] = create_10000_points(df["loc"][row][0],df["loc"][row][1],poly_New_Caledonia)
df.points[33] = create_10000_points_other(df["loc"][33][0],df["loc"][33][1],poly_New_Caledonia)
```

```python
df["cumsum"] = df.column_2.cumsum()
```
